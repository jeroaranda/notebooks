```python
import matplotlib.pyplot as plt
import matplotlib.collections as mcoll


```

# Graphics
## Bezier curves?

```python
n=30
plt.rcParams['figure.figsize'] = (10,11)
segments=[]
a=list(range(n))
for i in range(n-1):
     segments.append([[a[i],0],[0,a[n-i-1]]])
for i in range(n-1):
    segments.append([[-a[i],0],[0,a[n-i-1]]])
for i in range(n):
     segments.append([[a[i],0],[0,-a[n-i-1]]])
for i in range(n):
    segments.append([[-a[i],0],[0,-a[n-i-1]]])
lc = mcoll.LineCollection(segments,alpha=1,zorder=1,linewidth=2,color='Coral')

fig,ax=plt.subplots()
ax.add_collection(lc)
ax.set_xlim([-n,n])

ax.set_ylim([-n,n])
ax.axis('off')
fig.savefig('bezier.png',transparent=True)
```
